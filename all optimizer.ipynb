# Momentum

```python
v1 = 0
v2 = 0
gamma = 0.5
learning_rate = 0.4
x1 = -4
x2 = -6
l1_gd = []
l2_gd = []

for i in range(20):
    l1_gd.append(x1)
    l2_gd.append(x2)
    
    v1 = gamma * v1 + (0.2 * x1)
    v2 = gamma * v2 + (4 * x2)
    
    x1 = x1 - learning_rate * v1
    x2 = x2 = learning_rate * v2
    
```

# Adagrad 

```python
v1 = 0
v2 = 0
gamma = 0.9
learning_rate = 0.4
eps = np.linspace(0.1, 1)
x1 = -4
x2 = -6
l1_gd = []
l2_gd = []

for i in range(20):
    
    l1_gd.append(x1)
    l2_gd.append(x2)
    
    v1 = v1 + (0.2 * x1) **2
    v2 = v2 + (0.4 * x2) **2
    
    x1 = x1 - (learning_rate/np.sqrt(v1 + eps))*0.2*x1
    x2 = x2 - (learning_rate/np.sqrt(v2 + eps))*4*x2
```

# RMSprop

```python
v1 = 0
v2 = 0
gamma = 0.9
learning_rate = 0.4
eps = np.linspace(0.1, 1)
x1 = -4
x2 = -6
l1_gd = []
l2_gd = []

for i in range(20):
    l1_gd.append(x1)
    l2_gd.append(x2)
    
    v1 = gamma * v1 + (1-gamma) * (0.2 * x1)**2
    v2 = gamma * v2 + (1-gamma) * (4 * x2)**2
    
    x1 = x1 - (learning_rate/np.sqrt(v1 + eps))*0.2*x1
    x2 = x2 - (learning_rate/np.sqrt(v2 + eps))*0.2*x2
```

# Adam

```python
v1,v2,s1,s2=0,0,0,0
gamma,beta,lr=0.9,0.999,0.4
x1,x2=-6,-6
l1_adam,l2_adam=[],[]
c = np.linspace(0.001, 0.01)
for i in range(20):
    l1_adam.append(x1)
    l2_adam.append(x2)
    
    v1=gamma*v1+(1-gamma)*(0.2*x1)
    v2=gamma*v2+(1-gamma)*(4*x2)
    
    s1=beta*s1+(1-beta)*(0.2*x1)**2
    s2=beta*s2+(1-beta)*(4*x2)**2
    
    m_hat_v1= v1 / (1 - np.power(gamma, i+1))
    m_hat_v2 = v2 / (1 - np.power(gamma, i+1))
    
    s_hat_s1= s1 / (1 - np.power(beta, i+1))
    s_hat_s2 = s2 / (1 - np.power(beta, i+1))
    
    x1=x1-(lr)*(m_hat_v1/np.sqrt((s_hat_s1)+c))
    x2=x2-(lr)*(m_hat_v2/np.sqrt((s_hat_s2)+c))
```
